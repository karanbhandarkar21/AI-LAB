```python

#Matplotlib

import matplotlib.pyplot as plt
import numpy as np
subject = np.array(['A','B','C','D','E'])
marks = np.array([80,82,78,95,65])
plt.plot(subject, marks )
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("student performance")
plt.show()
```

```python
plt.plot(subject, marks )
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("student performance")
plt.grid()
plt.show()
```

```python
plt.plot(subject, marks )
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("student performance")
plt.bar(subject, marks)
plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.plot(subject, marks )
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("student performance")
plt.scatter(subject , marks )
plt.plot(marker= '*', lineStyles='dotted', color='orange')
plt.show()
```

```python
import matplotlib.pyplot as plt

subject = ['A', 'B', 'C' , 'D', 'E']
marks = [65,70,75,80,85]

plt.plot(subject, marks, marker='*', linestyle='dotted', color='orange')
plt.scatter(subject, marks)

plt.xlabel("Subject")
plt.ylabel("Marks")
plt.title("Student Performance")

plt.show()
```
